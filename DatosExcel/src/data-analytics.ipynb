# Analisis de Datos con Python y Pandas

### 1. Cargue y limpieza de Datos

```python
import pandas as pd
import datetime as dt

# Leer datos del archivo de datos en Excel (*.csv)
#df_source = pd.read_csv("../data/MiArchivo.csv", index_col=0, encoding='latin-1') #Se utiliza el index_col=0 para utilizar la primera columna del dataframe como indice
df_source = pd.read_csv("../data/MiArchivo.csv", encoding='latin-1')
df = df_source.copy() # Trabajar sobre una copia de los datos. No quiero danar los originales

# LIMPIEZA DE DATOS
# =================
# CONVERSIONES DE TIPOS Y DE FORMATO SOBRE LAS COLUMNAS DEL DATAFRAME
df['Salario Anual'] = df['Salario Anual'].astype(float)
df['Sueldo'] = df['Sueldo'].astype(float)
df['Fch comienzo'] = pd.to_numeric(df['Fch comienzo'], errors='coerce')
df['Fch nacimiento'] = pd.to_numeric(df['Fch nacimiento'], errors='coerce')
df['Fch comienzo'] = pd.TimedeltaIndex(df['Fch comienzo'], unit='d') + dt.datetime(1900,1,1) #Convertir fechas en formato "Excel" a formato "YYYY-mm-dd"
df['Fch nacimiento'] = pd.TimedeltaIndex(df['Fch nacimiento'], unit='d') + dt.datetime(1900,1,1) #Convertir fechas en formato "Excel" a formato "yyyy-mm-dd"
df['Apellido'] = df['Apellido'].str.upper() # Convertir apellidos a mayusculas
df['Nombre'] = df['Nombre'].str.upper() #Convertir nombres a mayusculas


# unique_values = df['Seccional'].unique() #Realizar un "unique" para obtener todos los valores unicos que contiene una columna
# filtered_df = df.query('Seccional == "0"') #Consultar todos los elementos de la columna "Seccional" que sean iguales a "0"
# total_salarios = df['Salario Anual'].sum() #Sumar todos los valores de la columna 'Salario Anual'
# print("Valor total de salarios: ", '${:0.2f}'.format(total_salarios))

# Agregar una fila de totales con la suma de todas las columnas al DataFrame 
df.loc[len(df)] = {
           'ID Empleado': 'TOTALES', 
           'Apellido': '', 
           'Nombre': '', 
           'Seccional': '', 
           'Facultad': '',
           'Salario Anual': df['Salario Anual'].sum(), 
           'Cargo':'', 
           'Fch comienzo': '0000-00-00', 
           'Fch nacimiento': '0000-00-00', 
           'Sueldo': df['Sueldo'].sum()
        }
# df['Total'] = df.sum(axis=1)  # Agregar columna de totales con la suma de todas las columnas al DataFrame (axis=1)


df
```
